# Demonstração KNN com Flor Iris

Base de Dados Iris

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Carregar o conjunto de dados Iris
iris = load_iris()

print("Amostra de entrada")
print(iris.data)
print()
print("Rótulo de Classe")
print(iris.target)
print()
print("Nome das Classes")
print(iris.target_names)
print()
print("Nome dos Atributos")
print(iris.feature_names)
print()
```

Separando o conjunto de dados em conjunto de treinamento e conjunto de teste

```python
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)

print ("x train")
print(x_train)

print ("x test")
print(x_test)

print("y train")
print(y_train)

print("y test")
print(y_test)
```

Treinando o modelo KNN usando o conjunto de treinamento

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
```

Avaliando o modelo usando o conjunto de test

```python
y_pred = knn.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Acurácia do modelo:", accuracy)
```

Efetuando previsões

```python
for i in range (1000):
    new_flower = np.array([[np.random.rand() * 10, np.random.rand() * 10, np.random.rand() * 10, np.random.rand() * 10]])
    prediction = knn.predict(new_flower)
    print(i, new_flower)
    print("Espécie da flor desconhecida:", iris.target_names[prediction])
    print()
    
```
